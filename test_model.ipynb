```python
# https://github.com/tensorflow/tensorflow/issues/22697
import requests
from PIL import Image
from io import BytesIO
import numpy
```

```python
# croissant = requests.get('http://www.paul.fr/38207-fiche_product/le-croissant.jpg')
croissant = requests.get('https://media-cdn.tripadvisor.com/media/photo-s/0e/59/3a/e2/le-pain-au-chocolat-a.jpg')
croissant = Image.open(BytesIO(croissant.content))
croissant.thumbnail((224,224))
croissant = croissant.resize((224,224))
croissant_np = numpy.expand_dims(numpy.array(croissant.convert('RGB')),0)
croissant
```

```python
from tensorflow.keras.models import load_model
```

```python
model = load_model('croissant.267.hdf5')
```

```python
prediction = model.predict(croissant_np)[0,0]
print(prediction)
if  prediction > 0.5:
    print('croissant')
else:
    print('pain au chocolat')
```
